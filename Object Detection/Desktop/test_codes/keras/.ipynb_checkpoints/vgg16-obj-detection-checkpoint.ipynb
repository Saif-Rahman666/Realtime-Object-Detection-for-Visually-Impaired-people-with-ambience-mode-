```python
import numpy as np 
import matplotlib.pyplot as plt
import glob
import cv2

from keras.models import Model, Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from keras.layers.normalization import BatchNormalization
import os
import seaborn as sns
from keras.applications.vgg16 import VGG16
import tensorflow as tf
tf.compat.v1.logging.set_verbosity(tf.compat.v1.logging.ERROR)


import matplotlib.gridspec as gridspec

from os.path import splitext,basename
from tensorflow.keras.models  import model_from_json
from keras.preprocessing.image import load_img, img_to_array
from keras.applications.mobilenet_v2 import preprocess_input
from sklearn.preprocessing import LabelEncoder
from tensorflow import keras
```

```python
# Read input images and assign labels based on folder names
print(os.listdir("E:/ambience_mode/dataset"))
```

```python
SIZE = 256  #Resize images
```

```python
#Capture training data and labels into respective lists
train_images = []
train_labels = []
label = os.listdir("E:/ambience_mode/dataset/train/")
i = 0
for directory_path in sorted(glob.glob("E:/ambience_mode/dataset/train/*")):
    if i<5:
        tlabel = label[i] 
    for img_path in glob.glob(os.path.join(directory_path, "*")):
        print(img_path)
        try :       
            img = cv2.imread(img_path, cv2.IMREAD_COLOR)       
            img = cv2.resize(img, (SIZE, SIZE))
            img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
            #img = np.array(img)
            #img = img/255.0
        except Exception as e :
            print("image error")
        train_images.append(img)
        train_labels.append(tlabel)
    i+=1
```

```python
#Convert lists to arrays
train_images = np.array(train_images)
train_labels = np.array(train_labels)
print(train_labels)
```

```python
# Capture test/validation data and labels into respective lists
test_images = []
test_labels = []
i = 0
print(label)
for directory_path in glob.glob("E:/ambience_mode/dataset/validation/*"):
    if i<5:
        tstlabel = label[i] 
    for img_path in glob.glob(os.path.join(directory_path, "*")):
        print(img_path)
        img = cv2.imread(img_path, cv2.IMREAD_COLOR)
        img = cv2.resize(img, (SIZE, SIZE))
        img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
        #img = np.array(img)
        #img = img/255.0
        test_images.append(img)
        test_labels.append(tstlabel)
    i+=1
        
print(test_labels)        
```

```python
#Convert lists to arrays 
test_images = np.array(test_images)
test_labels = np.array(test_labels)
```

```python
#Encode labels from text to integers
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(test_labels)
test_labels_encoded = le.transform(test_labels)
le.fit(train_labels)
train_labels_encoded = le.transform(train_labels)
```

```python
#Split data into test and train datasets (already split but assigning to meaningful convention)
x_train, y_train, x_test, y_test = train_images, train_labels_encoded, test_images, test_labels_encoded
```

```python
# Normalize pixel values to between 0 and 1
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
#One hot encode y values for neural network
from tensorflow.keras.utils import to_categorical
y_train_one_hot = to_categorical(y_train)
y_test_one_hot = to_categorical(y_test)
```

```python
#Load model wothout classifier/fully connected layers
VGG_model = VGG16(weights='imagenet', include_top=False, input_shape=(SIZE, SIZE, 3))
```

```python
VGG_model.summary()
```

```python
#Make loaded layers as non-trainable. This is important as we want to work with pre-trained weight
for layer in VGG_model.layers:
	layer.trainable = False
```

```python
VGG_model.summary()
```

```python
#Now, let us use features from convolutional network for RF
feature_extractor=VGG_model.predict(x_train)
```

```python
features = feature_extractor.reshape(feature_extractor.shape[0], -1)
```

```python
X_for_RF = features
```

```python
#RANDOM FOREST
from sklearn.ensemble import RandomForestClassifier
RF_model = RandomForestClassifier(n_estimators = 200, random_state = 60)
```

```python
# Train the model on training data
RF_model.fit(X_for_RF, y_train)
```

```python
#RF_model.model.save("E:/ambience_mode/model.h5")
VGG_model.save("E:/ambience_mode/model.h5")
```

```python
#Send test data through same feature extractor process
X_test_feature = VGG_model.predict(x_test)
X_test_features = X_test_feature.reshape(X_test_feature.shape[0], -1)
```

```python
#Now predict using the trained RF model. 
prediction_RF = RF_model.predict(X_test_features)
```

```python
#Inverse le transform to get original label back. 
prediction_RF = le.inverse_transform(prediction_RF)
```

```python
from sklearn.metrics import confusion_matrix
import pandas as pd
```

```python
cm_labels = np.unique(test_labels)
cm_array = confusion_matrix(test_labels, prediction_RF)
cm_array_df = pd.DataFrame(cm_array, index=cm_labels, columns=cm_labels)


#cm = confusion_matrix(test_labels, prediction_RF)
#print(cm)

%matplotlib inline

plt.figure(figsize=(10,7)).supxlabel('Figure : Confusion Matrix')

sns.heatmap(cm_array_df, annot=True,cbar = False) 
#sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
